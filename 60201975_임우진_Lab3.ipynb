```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getScore(): # 점수 생성
    Score = np.random.randint(100, size = (10, 4))
    return Score
```

```python
def Score_kor_eng(score): # 영어와 국어 점수의 산점도 그래프
    plt.scatter(score[:, 1], score[:, 0], marker = '^')
    plt.xlabel('English')
    plt.ylabel('Korean')
    plt.show()
```

```python
def Score_histogram_subplot(score): # 서브플롯 활용한 히스토그램

    plt.subplot(2, 2, 1)
    plt.hist(score[:, 0], bins=10)
    plt.xlabel('Korean')
    plt.ylabel('Count')

    plt.subplot(2, 2, 2)
    plt.hist(score[:, 1], bins=10)
    plt.xlabel('English')
    plt.ylabel('Count')

    plt.subplot(2, 2, 3)
    plt.hist(score[:, 2], bins=10)
    plt.xlabel('Math')
    plt.ylabel('Count')

    plt.subplot(2, 2, 4)
    plt.hist(score[:, 3], bins=10)
    plt.xlabel('Science')
    plt.ylabel('Count')

    plt.tight_layout(pad=3.0)  # 브플롯과 가장자리 사이의 여백

    plt.show()
```

```python
def Score_pie(scores): # 파이 그래프를 통한 비율 확인
    total_mean_Score = scores.mean(axis=1)
    print(total_mean_Score)
    
    A_student = np.sum(total_mean_Score >= 90)
    B_student = np.sum((total_mean_Score >= 80) & (total_mean_Score < 90))
    C_student = np.sum((total_mean_Score >= 70) & (total_mean_Score < 80))
    D_student = np.sum(total_mean_Score < 70)
    

    grade = [A_student, B_student, C_student, D_student]
    label = ['A', 'B', 'C', 'D']
    plt.pie(grade, labels = label, autopct='%.1f%%') # 소수점 1번째 자리로 제한
    plt.show()
```

```python
def Score_boxPlot(score): # 박스플롯 그래프
    plt.boxplot(score, labels= ['kor', 'eng', 'math', 'sci'])
```

```python
def main():
    score = getScore()
    print(score)
    Score_kor_eng(score)
    Score_histogram_subplot(score)
    Score_pie(score)
    Score_boxPlot(score)
```

```python
main()
```
